```python
import pandas as pd 
```

```python
df=pd.read_csv("imdb_top_1000.csv")
```

```python
df 
```

## remove unneseccary columns

```python
df = df.drop(['Gross', 'Meta_score','Star1','Star2','Star3','Star4'], axis=1)
```

```python
df = df.dropna()
```

```python
genre_counts = df['Genre'].value_counts()
```

```python
most_popular_genre = genre_counts.idxmax()
```

```python
print("The most popular genre is:", most_popular_genre)
```

```python
genre_counts_df = pd.DataFrame(genre_counts)
```

```python
genre_counts_df.columns = ['Count']
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
genre_counts_df.plot(kind='bar', legend=False)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre of Movies/TV Shows')
plt.show()
```

```python
director_counts = df['Director'].value_counts()
```

```python
director_counts_df = pd.DataFrame(director_counts)
```

```python
director_counts_df.columns = ['Count']
```

```python
director_counts_df.plot(kind='bar', legend=False)
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Director with the Most Top-Rated Movies/TV Shows')
plt.show()
```
